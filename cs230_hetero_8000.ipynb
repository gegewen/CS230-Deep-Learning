```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from numpy import genfromtxt
import scipy.io as sio
from PIL import Image
import random
import scipy
import imageio
from sklearn import preprocessing
%matplotlib inline 
```

```python
# load dataset from mat file
sat = sio.loadmat('/Users/gege/Downloads/hetero_8000.mat')
rad = sio.loadmat('/Users/gege/Downloads/hetero_8000_k_mat.mat')
X = rad['hetero_8000_k_mat'].T
Y = sat['all'].T
train_set_X = X[:768,]
train_set_Y = Y[:768,]
val_set_X = X[768:896, ]
val_set_Y = Y[768:896, ]
test_set_X = X[896:,]
test_set_Y = Y[896:,]
```

```python
# pad 50*50 image to 256*256 image, copy over three channels
def generateImage(dataset_X, dataset_Y, name):   
    for i in range(len(dataset_X)):
        mat_X = dataset_X[i].reshape(50,50)
        mat_X = np.interp(mat_X, (0, 2000), (0, 255)).astype(int)
    
        mat_X_padding = np.zeros((256, 256))
        mat_X_padding[104:104+mat_X.shape[0], 104:104+mat_X.shape[1]] = mat_X
    
        mat_Y = dataset_Y[i].reshape(50,50)
        mat_Y = np.interp(mat_Y, (0,1), (0, 255)).astype(int)

        mat_Y_padding = np.zeros((256, 256))
        mat_Y_padding[104:104+mat_Y.shape[0], 104:104+mat_Y.shape[1]] = mat_Y
    
        result = np.concatenate((mat_X_padding, mat_Y_padding), axis=1)
        test = np.dstack((result, np.dstack((result, result))))
    
        imageio.imwrite(name + '/' + str(i + 1) + '.jpg', test)
```

```python
def generateUnaligned(dataset_X, dataset_Y, name):   
    for i in range(len(dataset_X)):
        mat_X = dataset_X[i].reshape(50,50)
        mat_X = np.interp(mat_X, (0, 2000), (0, 255)).astype(int)
    
        mat_X_padding = np.zeros((256, 256))
        mat_X_padding[104:104+mat_X.shape[0], 104:104+mat_X.shape[1]] = mat_X
    
        mat_Y = dataset_Y[i].reshape(50,50)
        mat_Y = np.interp(mat_Y, (0,1), (0, 255)).astype(int)

        mat_Y_padding = np.zeros((256, 256))
        mat_Y_padding[104:104+mat_Y.shape[0], 104:104+mat_Y.shape[1]] = mat_Y
    
        test_X = np.dstack((mat_X_padding, np.dstack((mat_X_padding, mat_X_padding))))
        test_Y = np.dstack((mat_Y_padding, np.dstack((mat_Y_padding, mat_Y_padding))))

        imageio.imwrite(name + 'A/' + str(i + 1) + '_AB.jpg', test_X)
        imageio.imwrite(name + 'B/' + str(i + 1) + '_AB.jpg', test_Y)
```

```python
# generate iamges for train/validation/test set
generateImage(train_set_X, train_set_Y, 'train')
generateImage(val_set_X, val_set_Y, 'val')
generateImage(test_set_X, test_set_Y, 'test')
generateUnaligned(train_set_X, train_set_Y, 'train')
generateUnaligned(test_set_X, test_set_Y, 'test')
```
